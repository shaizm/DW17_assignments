### Registration & Login System

```python
# STEP 1

# Importing libraries to handel excel file.
import openpyxl
import pathlib
from os.path import exists

path = pathlib.Path.home() / 'Desktop\\Credentials.xlsx'
regstr = input('Do you wish to register or login ?\n')


# Collecting user_id  for registration or for login.
def ask_regi(x):
    global user_id
    if x.lower() == 'register':
        user_id = input('\nPlease enter your Email ID as username:\n')
        ask_email()
    
    elif x.lower() == 'login':
        user_id = input('\nPlease enter your Email ID as username:\n')
        compare_email()
        
    else:
        print("Invalid Input, Try Again")
        return             

# Checking if Email ID is valid   
def ask_email():
    if '@' and '.' not in user_id:
        print("Invalid Username: '@' or '.'  not found")
    else:
        check_email()

# Further checking for other parameters     
def check_email():
    global new_user_id
    index_1 = user_id.index('@')
    index_2 = user_id.index('.')
    new_user_id = user_id[index_1:]
        
    if new_user_id == user_id:
        print("Invalid Username: no characters found before '@'")
        
    elif not user_id[0].isalpha():
        print("Invalid Username: can not begin with numbers or special characters")
        
    elif index_2 == index_1+1:
        print("Invalid Usrname: '@' & '.' can not be consecutive characters")
        
    elif index_1 > index_2:
        print("Invalid Username: '@' should always preceed '.'")
        
    elif not user_id[0].isalpha():
        print("Invalid Username: no special characters or numbers at the beginning ")
        
    else:
        ask_password()

        
# Collecting password and verifying it        
def ask_password():
    global password
    print("\n***INSTRUCTIONS***")
    print("1. Password must have minimun 5 or maximum 16 characters.")
    print("2. Atleast one special character, one digit.")
    print("3. one uppercase & one lowercase character.")
    password = input("\nEnter your password:\n")
     
    if len(password)<5 or len(password)>16:
        print("Invalid Password: Must have min 5 or max 16 characters")
        ask_password()

    elif password.isalnum():
        print("Invalid Password: Must have atleast 1 special character")
        ask_password()

    elif not any(i.isdigit() for i in password):
        print("Invalid Password: Must contain atleast 1 digit")
        ask_password()

    elif not any(i.isupper() for i in password):
        print("Invalid Password: Must contain atleast 1 capital letter")
        ask_password()

    elif not any(i.islower() for i in password):
        print("Invalid Password: Must contain atleast 1 small letter")
        ask_password()
    
    else:
        storing_credential()
        print("\nPassword Saved")
        print("your registration is done.")

        
# Step  2
# Saving all previous and current username and password in excel file
def storing_credential():
    if not exists(path):  
        workbook = openpyxl.Workbook()
        sheet = workbook.active

        c1 = sheet.cell(row = 1, column = 1)
        c1.value = "User Names"
        c2 = sheet.cell(row = 1, column = 2)
        c2.value = "Passwords"

        workbook.save(filename = path)
        workbook.close()

    if exists(path):
        workbook = openpyxl.load_workbook(path)
        sheet = workbook.active
        data = (user_id, password)
        sheet.append(data)

        workbook.save(filename = path)
        workbook.close()


# Step  3
# Checking if username already exists and also retrieving respective passwords
def compare_email():
    workbook = openpyxl.load_workbook(path)
    sheet = workbook.active
    userids = []
    count = 0
    
    no_of_rows = sheet.max_row
    
    for i in range(2, no_of_rows+1):
        userid = sheet.cell(i,1)
        userids.append(userid.value)
        
    for i in userids:
        if i == user_id:
            index_ = userids.index(i) + 2
            password = input("\nEnter your password:\n")
            paswrd = sheet.cell(index_, 2)
        
            if (paswrd.value) == password:
                print("Login Successful")
        
            else:
                print("Password incorrect")
                forgot = input("Retrieve Password, yes/no\n")
                if forgot.lower() == "yes":
                    print("Your password is: ", paswrd.value)
        
        else:
            count = count + 1
            if count >= len(userids):
                print("\nUser name not found, please register")     
    
    workbook.save(filename = path)
    workbook.close()
    
    
    
ask_regi(regstr)
```
